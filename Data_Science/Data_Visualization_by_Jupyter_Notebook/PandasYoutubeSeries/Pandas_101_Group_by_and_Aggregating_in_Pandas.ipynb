```python
import pandas as pd

df = pd.read_csv('Flavors.csv')

df
```

```python
from os import getgroups
#Group by Base Flavor
list_feature = ['Flavor Rating', 'Texture Rating','Total Rating']
group_by_frame = df.groupby('Base Flavor')[list_feature]

group_by_frame

group_by_frame_mean = group_by_frame.mean()

group_by_frame_mean
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor')[list_feature].sum()
```

```python
agg_funs = ['mean','max','sum','count']
list_feature = ['Flavor Rating', 'Texture Rating','Total Rating']
df.groupby('Base Flavor').agg({x: agg_funs for x in list_feature})
```

```python
df.groupby(['Base Flavor','Liked']).agg({list_feature[0]: agg_funs})
```
