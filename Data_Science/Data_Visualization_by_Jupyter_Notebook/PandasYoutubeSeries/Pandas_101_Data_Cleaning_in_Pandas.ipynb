```python
import pandas as pd

df = pd.read_excel('Customer Call List.xlsx')
display(df.head())
```

```python
#Dropping duplicates data and columns
df = df.drop_duplicates()
df = df.drop(columns = 'Not_Useful_Column')
#Cleaning Last_Name Column
df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
#Cleaning Phone_Number_Column
df['Phone_Number'] = df['Phone_Number'].astype(str) # Transform Phone_Number from float to string
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
#Creating columns from Address column
df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(',', n = 2,expand = True)
#Replace from Yes to Y and from No to N in Do_not_contact columns
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
#Standardize Nan by filna()
df = df.replace('N/a','')
df = df.replace('Nan','')
df = df.fillna('')
#Drop value 'Y' in Do_Not_Contact column and drop na in dataframe
"""for x in df.index:
    if df.loc[x, "Do_Not_Contact"]=='Y':
      df.drop(x,inplace = True)"""
df = df[df['Do_Not_Contact'] !='Y']
df.dropna(subset='Phone_Number', inplace= True)
#Reset index in dataframe
df = df.reset_index(drop = True)
#Display new dataframe
display(df)
```
