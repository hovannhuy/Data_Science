```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Ice Cream Ratings.csv', index_col = 'Date')
df
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings by Date', xlabel = 'Daily Ratings', ylabel= 'Scores')
```

```python
df.plot.barh(stacked = True, title = 'Ice Cream Ratings by Date')
```

```python
df.plot.scatter(x = 'Texture Rating', y='Overall Rating', s = df['Flavor Rating']*1000, c = 'Yellow')
```

```python
df.plot.hist(bins= 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (20,10))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,10))
```
